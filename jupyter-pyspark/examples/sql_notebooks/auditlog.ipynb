```python
%%sql

SELECT
    *
FROM 
    "vast-audit-log-bucket|vast_audit_log_schema".vast_audit_log_table
WHERE
    view_path like '/csnow%' 
    and protocol <> 'NDB'
ORDER BY
    time DESC
LIMIT 10
```
